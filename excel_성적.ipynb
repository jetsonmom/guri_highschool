코랩에서 파이선으로  학생5명의 이름과 국어, 수학,과학 성적을 입력하고 평균, 합계, 전체 과목별 각각 평균 값을  나타내는 엑셀처럼 도표를 만드는 코드를 만들어줘

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# 학생 데이터 입력
students = []
print("=== 학생 성적 입력 ===")
for i in range(5):
    print(f"\n{i+1}번째 학생 정보 입력:")
    name = input("이름: ")
    korean = int(input("국어 점수: "))
    math = int(input("수학 점수: "))
    science = int(input("과학 점수: "))

    students.append({
        '이름': name,
        '국어': korean,
        '수학': math,
        '과학': science
    })

# DataFrame 생성
df = pd.DataFrame(students)

# 개인별 합계 및 평균 계산
df['합계'] = df[['국어', '수학', '과학']].sum(axis=1)
df['평균'] = df[['국어', '수학', '과학']].mean(axis=1).round(2)

# 과목별 평균 행 추가
subject_avg = df[['국어', '수학', '과학']].mean().round(2)
avg_row = pd.DataFrame([{
    '이름': '과목별 평균',
    '국어': subject_avg['국어'],
    '수학': subject_avg['수학'],
    '과학': subject_avg['과학'],
    '합계': '-',
    '평균': '-'
}])

df_display = pd.concat([df, avg_row], ignore_index=True)

# 콘솔에 표 출력
print("\n" + "="*60)
print("                    학생 성적표")
print("="*60)
print(df_display.to_string(index=False))
print("="*60)

# 시각적 테이블 생성 (엑셀 스타일)
fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('tight')
ax.axis('off')

# 테이블 데이터 준비
table_data = []
table_data.append(['이름', '국어', '수학', '과학', '합계', '평균'])  # 헤더

# 학생 데이터
for idx, row in df.iterrows():
    table_data.append([
        row['이름'],
        int(row['국어']),
        int(row['수학']),
        int(row['과학']),
        int(row['합계']),
        f"{row['평균']:.2f}"
    ])

# 과목별 평균 행
table_data.append([
    '과목별 평균',
    f"{subject_avg['국어']:.2f}",
    f"{subject_avg['수학']:.2f}",
    f"{subject_avg['과학']:.2f}",
    '-',
    '-'
])

# 테이블 생성
table = ax.table(cellText=table_data, cellLoc='center', loc='center',
                colWidths=[0.2, 0.15, 0.15, 0.15, 0.15, 0.15])

# 테이블 스타일링
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 2.5)

# 헤더 스타일 (첫 번째 행)
for i in range(6):
    cell = table[(0, i)]
    cell.set_facecolor('#4472C4')
    cell.set_text_props(weight='bold', color='white')

# 데이터 행 스타일 (교차 색상)
for i in range(1, len(table_data)):
    for j in range(6):
        cell = table[(i, j)]
        if i == len(table_data) - 1:  # 마지막 행 (과목별 평균)
            cell.set_facecolor('#F4B084')
            cell.set_text_props(weight='bold')
        elif i % 2 == 0:
            cell.set_facecolor('#E7E6E6')
        else:
            cell.set_facecolor('#FFFFFF')

        # 테두리
        cell.set_edgecolor('#000000')
        cell.set_linewidth(1)

plt.title('학생 성적 통계표', fontsize=16, weight='bold', pad=20)
plt.show()

# 추가 통계
print(f"\n전체 학생 평균: {df['평균'].mean():.2f}점")
print(f"최고 점수: {df['이름'].iloc[df['평균'].idxmax()]} ({df['평균'].max():.2f}점)")
print(f"최저 점수: {df['이름'].iloc[df['평균'].idxmin()]} ({df['평균'].min():.2f}점)")
```
